```python
# http请求
import urllib.request
import urllib.parse
params = urllib.parse.urlencode({'catId': 0, 'initiative_id': 'SB_20200721001233', 'SearchText': 'M.2+SSD+case'})
url = "https://www.aliexpress.com/wholesale?%s" % params
with urllib.request.urlopen(url) as f:
    html = f.read().decode('')
```

```python
# 解析html
from bs4 import BeautifulSoup
soup = BeautifulSoup(html)
soup.prettify
```
