```python
# 处理结果数据
df1 = pd.read_excel('公交od1.xlsx')
df2 = pd.read_excel('公交od2.xlsx')
df3 = pd.read_excel('drivingod1.xlsx')
df4 = pd.read_excel('drivingod2.xlsx')
df5 = pd.read_excel('walkingod1.xlsx')
df6 = pd.read_excel('walkingod2.xlsx')
```

```python
od1 = df1.merge(df3, how='inner', on=['出发点','目的地'])\
        .merge(df5, how='inner', on=['出发点','目的地'])
od2 = df2.merge(df4, how='inner', on=['出发点','目的地'])\
        .merge(df6, how='inner', on=['出发点','目的地'])
od = pd.concat([od1, od2])

res_df = od[['出发点','origin_x','目的地','dest_x','公交时间','duration_distance_x','duration_distance_y']]
res_df.columns = ['出发点','origin','目的地','destination','公交时间','驾车时间距离','步行时间距离']
res_df.set_index('出发点').to_excel('result.xlsx')
res_df
```
