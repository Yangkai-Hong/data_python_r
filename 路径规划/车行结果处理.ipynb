```python
import pandas as pd
import numpy as np
```

```python
res = pd.DataFrame()
for i in range(1,26):
    file_name = 'fp' + str(i) + '.xlsx'
    df = pd.read_excel(file_name)[['居民点id','居民点location','设施点id','设施点location','duration_distance']]
    res = pd.concat([res, df],ignore_index=True)
res
```

```python
for i in range(1, 23):
    file_name = 'sp' + str(i) + '.xlsx'
    df = pd.read_excel(file_name)[['居民点id','居民点location','设施点id','设施点location','duration_distance']]
    res = pd.concat([res, df],ignore_index=True)
res
```

```python
# 补全数据
rest = pd.DataFrame()

df1 = pd.read_excel('undo1_driving.xlsx')[['居民点id','居民点location','设施点id','设施点location','duration_distance']]
df2 = pd.read_excel('undo2_driving.xlsx')[['居民点id','居民点location','设施点id','设施点location','duration_distance']]

rest = pd.concat([rest, df1, df2], ignore_index=True)
rest
```

```python
# 检查是否有缺失数据
done = res[res['duration_distance']!=',']
total_df = pd.concat([done, rest], ignore_index=True)
total_df
```

```python
ssd_79 = pd.read_excel('ssd79_driving.xlsx')[['居民点id','居民点location','设施点id','设施点location','duration_distance']]
ssd_79
```

```python
# 第一步：在130万行OD矩阵骑行.xsl 中用第49个医疗点的OD矩阵替换到第80个医疗点的OD矩阵，并把第49个医疗点的OD矩阵删掉，原有序号不变（没有第49个医疗点）
# 第二步:第79个医疗点ID改为120.51244，32.137315计算骑行OD矩阵，并替换原有第79个医疗点的OD矩阵
result = total_df[~total_df['设施点id'].isin([79,80])]
result = pd.concat([result, ssd_79], ignore_index=True)
result['设施点id'] = result['设施点id'].replace({49:80})

result['duration'] = pd.to_numeric(result['duration_distance'].str.split(',').str[0])
result['distance'] = pd.to_numeric(result['duration_distance'].str.split(',').str[1])
result
```

```python
result['设施点id'].nunique()
```

```python
result.set_index('居民点id').to_csv('./driving/车行OD矩阵.csv')
```

```python
# 最小时间
min_time = result.sort_values(by='duration')\
            .groupby(by='居民点id')\
            .head(1)\
            .sort_values(by='居民点id')
min_time[['居民点id','居民点location','设施点id','设施点location','duration']].set_index('居民点id').to_excel('./driving/车行OD矩阵最小时间.xlsx')
min_time
```

```python
# 最小距离
min_distance = result.sort_values(by='distance')\
            .groupby(by='居民点id')\
            .head(1)\
            .sort_values(by='居民点id')
min_distance[['居民点id','居民点location','设施点id','设施点location','distance']].set_index('居民点id').to_excel('./driving/车行OD矩阵最小距离.xlsx')
min_distance
```
