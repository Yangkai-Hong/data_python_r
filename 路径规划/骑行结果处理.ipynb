```python
import pandas as pd
import numpy as np
```

```python
res = pd.DataFrame()
for i in range(1,15):
    file_name = 'p' + str(i) + '.xlsx'
    try:
        df = pd.read_excel(file_name)[['居民点id','居民点location','设施点id','设施点location','duration_distance_bicycling']]
        res = pd.concat([res, df],ignore_index=True)
    except:
        print(file_name)
res
```

```python
done = res[res['duration_distance_bicycling']!=',']
undo = pd.read_excel('undo.xlsx')[['居民点id','居民点location','设施点id','设施点location','duration_distance_bicycling']]
total_df = pd.concat([done, undo], ignore_index=True)
total_df
```

```python
ssd_79 = pd.read_excel('ssd79.xlsx')[['居民点id','居民点location','设施点id','设施点location','duration_distance_bicycling']]
ssd_79['设施点id'] = 78
ssd_78 = ssd_79
ssd_78
```

```python
#第一步：在39万行OD矩阵车行.xsl 中用第48个医疗点的OD矩阵替换到第79个医疗点的OD矩阵，并把第48个医疗点的OD矩阵删掉，原有序号不变（没有第48个医疗点）
#第二步：第78个医疗点ID改为120.51244，32.137315计算车行OD矩阵，并替换原有第78个医疗点的OD矩阵
result = total_df[~total_df['设施点id'].isin([79,78])]
result = pd.concat([result, ssd_78], ignore_index=True)

result['设施点id'] = result['设施点id'].replace({48:79})
```

```python
result['duration'] = pd.to_numeric(result['duration_distance_bicycling'].str.split(',').str[0])
result['distance'] = pd.to_numeric(result['duration_distance_bicycling'].str.split(',').str[1])
```

```python
result['设施点id'].unique()
```

```python
result[result['duration_distance_bicycling']==',']
```

```python
result.set_index('居民点id').to_csv('./骑行OD矩阵.csv')
```

```python
# 最小时间
min_time = result.sort_values(by='duration')\
            .groupby(by='居民点id')\
            .head(1)\
            .sort_values(by='居民点id')
min_time[['居民点id','居民点location','设施点id','设施点location','duration']].set_index('居民点id').to_excel('./骑行OD矩阵最小时间.xlsx')
```

```python
# 最小距离
min_distance = result.sort_values(by='distance')\
            .groupby(by='居民点id')\
            .head(1)\
            .sort_values(by='居民点id')
min_distance[['居民点id','居民点location','设施点id','设施点location','distance']].set_index('居民点id').to_excel('./骑行OD矩阵最小距离.xlsx')
```
